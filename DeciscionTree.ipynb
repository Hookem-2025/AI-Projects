```
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn as sk
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #hehe
from sklearn.ensemble import RandomForestClassifier



train = pd.read_csv("train.csv", ) 
test = pd.read_csv("test.csv",index_col='id')


features = train.columns[:11]

#feature_cols = ['age', 'gender', 'height', 'weight','ap_hi','ap_lo','cholesterol', 'gluc','smoke','alco','active']


# load dataset
coolguy = pd.read_csv("train.csv", names = features, header=None)
X = train[features] # Features
y = train.cardio

#depth of 28?
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1) # 70% training and 30% test


random_forest = RandomForestClassifier(n_estimators=30, max_depth=13, random_state=1)


# Train Decision Tree Classifer
clf = random_forest.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

from sklearn.metrics import confusion_matrix

pd.DataFrame(
    confusion_matrix(y_test, y_pred),
    columns=['Predicted Not Cardio', 'Predicted Cardio'],
    index=['True Not Cardio', 'True Cardio']
)
```
